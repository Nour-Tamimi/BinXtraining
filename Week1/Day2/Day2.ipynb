```python

def calculate(numbers_list):
    Min=min(numbers_list)
    Max=max(numbers_list)
    Mean=sum(numbers_list)/len(numbers_list)
   
    print('The min',Min,' The max',Max,' The mean',Mean )
    
```

```python
numbers=[1,2,3,4,5]
calculate(numbers)
```

This is a function that calculates the min , max and mean (average for a list of numbers)

```python
class DataRecord:
    def __init__(self,a,b):
        self.a=a
        self.b=b
  
    def count(self):
        return self.a+self.b
```

```python
d=DataRecord(1,10)
print(d.count())
```

This is a class called datarecord tht has 2 attributes a and b and in has one method count that sum these two values
