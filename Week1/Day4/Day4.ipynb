```python
road=pd.read_csv('C:/Users/Lenovo/Downloads/automobile_dataset.csv')
```

```python
print(road.info())
```

```python
road.drop_duplicates(inplace=True)
```

```python
print(road.head())
```

```python
print(road.describe())
```

```python
road["Horsepower"].mean()
```

```python
road.isnull().sum()
```

```python
road["Fuel_Efficiency"] = road["Fuel_Efficiency"].fillna(road["Fuel_Efficiency"].mean())
```

```python
road.isnull().sum()
```

```python
road["Accident_History"]=road["Accident_History"].mean()
```

```python
road["Transmission"].mode()
```

```python
road["Transmission"]=road["Transmission"].fillna(road["Transmission"].mode()[0])
```

```python
road.isnull().sum()
```

```python
road.dropna(inplace=True)
```

first i dropped all the duplicates next i filled all the fuel efficiency feild with  the mean then i corrected it and filled only the null then i did the same for accident history field then i got the mode[0] for the transmission field and filled the null with it then i dropped everything else still null (service history,color,engine size,location)
